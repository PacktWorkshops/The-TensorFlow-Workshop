```python
import pandas as pd
```

```python
df = pd.read_csv('../Datasets/Bias_correction_ucl.csv')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
year_dummies = pd.get_dummies(df['Date'].dt.year, prefix='year')
year_dummies
```

```python
month_dummies = pd.get_dummies(df['Date'].dt.month, prefix='month')
month_dummies
```

```python
df2 = pd.concat([df, month_dummies, year_dummies], axis=1)
```

```python
df2.drop('Date', axis=1, inplace=True)
```
